```python
# Sorting functions

#################################
# Sorting - while loop, with "in_order" test function

def sortList1(x):
    x = list(x)
    while not in_order(x):
        for i in range(len(x)-1):
            if x[i] > x[i+1]:
                tmp = x[i]
                x[i] = x[i+1]
                x[i+1] = tmp
    return x
    
def in_order(x):
    for i in range(len(x)-1):
        if x[i] > x[i+1]:
            return False
    return True

#################################
# Sorting - while loop, with boolean for whether swaps made

def sortList2(x):
    x = list(x)
    sorting = True
    while sorting:
        sorting = False
        for i in range(len(x)-1):
            if x[i] > x[i+1]:
                tmp = x[i]
                x[i] = x[i+1]
                x[i+1] = tmp
                sorting = True
    return x

#################################
# Bubblesort

def bubble(x):
    x = list(x)
    for j in range(len(x)-1):
        for i in range(len(x)-j-1):
            if x[i] > x[i+1]:
                tmp = x[i]
                x[i] = x[i+1]
                x[i+1] = tmp
    return x
```

```python
# Colour images as 3D arrays 
# First effect, the spirit chick, is created by using first method, addressing individual intensities
# Second effect, the horror chick, is created by using second method, addressing pixel triples

%matplotlib inline
import pylab as pyl

img = pyl.imread('images/chick.png')
(d1,d2,d3) = img.shape

pyl.figure()
pyl.imshow(img)

############################################
# First effect (ghost chick):

img1 = pyl.array(img) # copy img as img1

for i in range(d1):
    for j in range(d2):
        for k in range(d3):
            img1[i,j,k] = 1 - img1[i,j,k]

pyl.figure()
pyl.imshow(img1)

############################################
# Second effect (horror chick):

img2 = pyl.array(img) # copy img as img2

for i in range(d1):
    for j in range(d2):
        if sum(img2[i,j]) < 1.5:
            img2[i,j] = (.0,.0,.0)

pyl.figure()
pyl.imshow(img2)

############################################
# Finally, show all the images
pyl.show()
```

```python
# Che Guevara task
# For each effect, the image is copied and all the modifications are made on the copy, not in the original image

img = pyl.imread('images/che.png')
(d1,d2,d3) = img.shape

pyl.figure()
pyl.imshow(img)

############################################
# First effect (strongly monochrome):

img1 = pyl.array(img) # copy image array

for i in range(d1):
    for j in range(d2):
        for k in range(d3):
            if img1[i,j,k] < 0.5:
                img1[i,j,k] = 0.0
            else:
                img1[i,j,k] = 1.0

pyl.figure()
pyl.imshow(img1)

##############################################
# Second effect (red/black):

img2 = pyl.array(img) # copy image array

for i in range(d1):
    for j in range(d2):
        if img2[i,j,0] < 0.5:
            img2[i,j] = (.0,.0,.0)
        else:
            img2[i,j] = (1.,.0,.0)

pyl.figure()
pyl.imshow(img2)

##############################################
# Third effect (face white, background red):

img3 = pyl.array(img) # copy image array

for i in range(d1):
    for j in range(d2):
        if img3[i,j,0] < 0.5:
            img3[i,j] = (.0,.0,.0) # black
        elif 55 < i < 160 and 55 < j < 140:
            img3[i,j] = (1.,1.,1.) # white
        else:
            img3[i,j] = (1.,.0,.0) # red

pyl.figure()
pyl.imshow(img3)

############################################
# Fourth effect (psychedelic):

img4 = pyl.array(img) # copy image array

for i in range(d1):
    for j in range(d2):
        if img4[i,j,0] > 0.66:
            img4[i,j] = (1.,.0,.0) # red
        elif img4[i,j,0] < 0.33:
            img4[i,j] = (.0,.0,1.) # blue
        else:
            img4[i,j] = (.0,1.,.0) # green

pyl.figure()
pyl.imshow(img4)

############################################
# Fifth effect (overlay chick image):

img5 = pyl.array(img) # copy image array
chick = pyl.imread('images/chick.png')
(c1,c2,c3) = chick.shape

(dx,dy) = (60,58) # displacement of chick image from top left
                  # corner, as components in x,y directions

for i in range(c1):
    for j in range(c2):
        img5[i+dy,j+dx] = chick[i,j]

pyl.figure()
pyl.imshow(img5)

############################################
# Finally, show all the images
pyl.show()
```
