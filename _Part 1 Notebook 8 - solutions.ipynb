```python
# Exercise 1 
# Sum of reciprocals function

def sum_of_reciprocals(N):
    sum_re = 0
    for x in range(1,N+1):
        sum_re += 1/x
    return (sum_re)

# Test calls for the function
print (sum_of_reciprocals(2))
print (sum_of_reciprocals(5))
print (sum_of_reciprocals(10))
```

```python
# Exercise 2
# Sum of corresponding elements from lists of equal lengths

def add_numlists(nums1,nums2):
    if len(nums1) == len(nums2):
        sums = []
        for i in range(len(nums1)):
            sums.append(nums1[i]+nums2[i])
        return (sums)
    else:
        return (None)
    
# Test calls for the function
print(add_numlists([3,1,5],[2,6,4]))
print(add_numlists([31,1,5,-1],[2,6,4]))
print(add_numlists([31,1,15,-1],[2,3,4,1]))
```

```python
# Exercise 3
# Determines if all the elements of a list are positive

def pos_numlist(nums):
    if all(elem >= 0 for elem in nums):
        return (True)
    else:
        return (False)

# Test calls for the function    
print(pos_numlist([3,1,9]))
print(pos_numlist([0,1,9]))
print(pos_numlist([3,-1,9]))
```

```python
# Exercise 4
# Distance between two asteroids
%matplotlib inline
import pylab as pyl

# Some alternative versions ...

########################################
# Simple version, where x and y vales assigned to
# separate explicit x and y variables

ast1_x = -150.4
ast1_y = 200.5

vel1_x = 4.9
vel1_y = -7.1

ast2_x = 122.6
ast2_y = 64.0

vel2_x = -5.2
vel2_y = -2.95

distances = []
times = range(51)

for t in times:
    d = ((ast1_x - ast2_x) ** 2 + (ast1_y - ast2_y) ** 2) ** 0.5
    distances.append(d)
    ast1_x += vel1_x
    ast1_y += vel1_y
    ast2_x += vel2_x
    ast2_y += vel2_y

pyl.plot(distances,'b.')
pyl.xlabel("Time (in hours)")
pyl.ylabel("Distance (in '000 of km)")
pyl.show()

########################################
# Version where x and y vales are grouped as XY pairs
# (with x and y accessed via index 0 and 1)

ast1 = [-150.4, 200.5]
vel1 = [4.9, -7.1]

ast2 = [122.6, 64.0]
vel2 = [-5.2, -2.95]

distances = []
times = range(51)

for t in times:
    d = ((ast1[0] - ast2[0]) ** 2 + (ast1[1] - ast2[1]) ** 2) ** 0.5
    distances.append(d)
    ast1[0] += vel1[0]
    ast1[1] += vel1[1]
    ast2[0] += vel2[0]
    ast2[1] += vel2[1]

pyl.plot(distances,'r.')
pyl.xlabel("Time (in hours)")
pyl.ylabel("Distance (in '000 of km)")
pyl.show()

########################################
# This version allows plotting for any time interval,
# not just complete (integer) hours. Has a function
# to compute current position, based on initial position
# velocity and time.

ast1 = [-150.4, 200.5]
vel1 = [4.9, -7.1]

ast2 = [122.6, 64.0]
vel2 = [-5.2, -2.95]

def posn_at_time(pos,vel,time):
    return ([pos[0] + vel[0] * time, pos[1] + vel[1] * time])

distances = []
times = range(50)
times = pyl.arange(0,50,0.5) # time points at 0.5 hour intervals

for t in times:
    pos1 = posn_at_time(ast1,vel1,t)
    pos2 = posn_at_time(ast2,vel2,t)
    d = ((pos1[0] - pos2[0]) ** 2 + (pos1[1] - pos2[1]) ** 2) ** 0.5
    distances.append(d)

pyl.plot(distances,'g.')
pyl.xlabel("Time (in hours)")
pyl.ylabel("Distance (in '000 of km)")
pyl.show()
```
